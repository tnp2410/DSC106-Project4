```python
import pandas as pd
import numpy as np
```

```python
pokemon = pd.read_csv('pokemon.csv')
pokemon
```

```python
pokemon.columns
```

```python
pokemon = pokemon.drop(columns = ['percentage_male', 'weight_kg', 'japanese_name', 'base_egg_steps'
, 'base_happiness', 'base_total', 'height_m', 'hp', 'is_legendary', 'generation',
                                  'experience_growth', 'capture_rate', 'sp_attack',
                                  'sp_defense', 'speed', 'attack', 'defense','abilities'])
pokemon
```

```python
pokemon['type'] = pokemon[['type1', 'type2']].values.tolist()
pokemon
```

```python
pokemon['type'] = pokemon['type'].apply(lambda x: [i for i in x if pd.notna(i)])
pokemon.drop(columns=['type1', 'type2'], inplace=True)
```

```python
pokemon
```

```python
pokemon.to_csv()
```
